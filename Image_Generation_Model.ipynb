# Import Packages

```python
import time
import keras_cv
import keras
import matplotlib.pyplot as plt
```

# construct a model

```python
model = keras_cv.models.StableDiffusion(
    img_width=512, img_height=512, jit_compile=False
)
```

# give it a prompt

```python
images = model.text_to_image("photograph of elon musk and mark zuckerberg mma fight on mars", batch_size=3)


def plot_images(images):
    plt.figure(figsize=(20, 20))
    for i in range(len(images)):
        ax = plt.subplot(1, len(images), i + 1)
        plt.imshow(images[i])
        plt.axis("off")


plot_images(images)
```

# more complex prompt

```python
images = model.text_to_image(
    "Portrait of a Sci-Fi man holding flamethrowers on the space station, "
    "cinematic lighting, high-resolution 3D art,, "
    "concept art, character concepts, digital painting, mystery, adventure",
    batch_size=3,
)
plot_images(images)
```
